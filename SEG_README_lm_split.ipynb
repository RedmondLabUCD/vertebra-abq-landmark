```python
from utils.process_predictions import superimpose
from torchvision.datasets.utils import list_files
import pandas as pd

root = os.path.abspath(os.path.join(os.getcwd(), os.pardir))

ops = [["UNet_FemHead",""],["UNet_FemHead","_AUG"],["UNet_ROI_FemHead"," Pred ROI"],["UNet_ROI_FemHead","_AUG Pred ROI"]]

modelex_1 = str(ops[0][0]+ops[0][1])
modelex_2 = str(ops[1][0]+ops[1][1])
modelex_3 = str(ops[2][0]+"_Pred")
modelex_4 = str(ops[3][0]+"_Pred_AUG")

superimposed_dir_1 = os.path.join(root,"Superimposed",modelex_1)
superimposed_dir_2 = os.path.join(root,"Superimposed",modelex_2)
superimposed_dir_3 = os.path.join(root,"Superimposed",modelex_3)
superimposed_dir_4 = os.path.join(root,"Superimposed",modelex_4)

if not os.path.exists(superimposed_dir_1): os.makedirs(superimposed_dir_1)
if not os.path.exists(superimposed_dir_2): os.makedirs(superimposed_dir_2)
if not os.path.exists(superimposed_dir_3): os.makedirs(superimposed_dir_3)
if not os.path.exists(superimposed_dir_4): os.makedirs(superimposed_dir_4) 
        
for fold in range(1,11):

    short_dir = os.path.join(root,"Dataset","Fold "+str(fold),"FemHead Masks")
    filenames = list_files(short_dir,".png",prefix=False)
    
    for file in filenames:
        filename = file[:-4]
        centroid_stats = pd.read_csv(os.path.join(root,"Results","Statistics","centroid_stats.csv"))

        dist_r, dist_l = superimpose(filename,superimposed_dir_1,fold,extra=ops[0][1],model=ops[0][0],roi=0,labels=["Target","UNet-Baseline"])
        centroid_stats.loc[centroid_stats["Image"]==int(filename),"Fold"] = fold
        if dist_r != -1:
            centroid_stats.loc[centroid_stats["Image"]==int(filename),str(modelex_1+"_Right")] = dist_r
            centroid_stats.loc[centroid_stats["Image"]==int(filename),str(modelex_1+"_Left")] = dist_l
            centroid_stats.loc[centroid_stats["Image"]==int(filename),str(modelex_1+"_Mean")] = (dist_l + dist_r)/2
        else:
            print(filename)

        dist_r, dist_l = superimpose(filename,superimposed_dir_2,fold,extra=ops[1][1],model=ops[1][0],roi=0,labels=["Target","UNet-AUG"])
        if dist_r != -1:
            centroid_stats.loc[centroid_stats["Image"]==int(filename),str(modelex_2+"_Right")] = dist_r
            centroid_stats.loc[centroid_stats["Image"]==int(filename),str(modelex_2+"_Left")] = dist_l
            centroid_stats.loc[centroid_stats["Image"]==int(filename),str(modelex_2+"_Mean")] = (dist_l + dist_r)/2

        if filename != "33914546":
            dist_r, dist_l = superimpose(filename,superimposed_dir_3,fold,extra=ops[2][1],model=ops[2][0],roi=0,labels=["Target","UNet-ROI"])
            if dist_r != -1:
                centroid_stats.loc[centroid_stats["Image"]==int(filename),str(modelex_3+"_Right")] = dist_r
                centroid_stats.loc[centroid_stats["Image"]==int(filename),str(modelex_3+"_Left")] = dist_l
                centroid_stats.loc[centroid_stats["Image"]==int(filename),str(modelex_3+"_Mean")] = (dist_l + dist_r)/2
                
        if filename != "33914546":
            dist_r, dist_l = superimpose(filename,superimposed_dir_4,fold,extra=ops[3][1],model=ops[3][0],roi=0,labels=["Target","UNet-ROI-AUG"])
            if dist_r != -1:
                centroid_stats.loc[centroid_stats["Image"]==int(filename),str(modelex_4+"_Right")] = dist_r
                centroid_stats.loc[centroid_stats["Image"]==int(filename),str(modelex_4+"_Left")] = dist_l
                centroid_stats.loc[centroid_stats["Image"]==int(filename),str(modelex_4+"_Mean")] = (dist_l + dist_r)/2

        centroid_stats.to_csv(os.path.join(root,"Results","Statistics",'centroid_stats.csv'),index=False)
```

```python
import pandas as pd
from utils.data_prep import rms_contrast
import numpy as np
import matplotlib.pyplot as plt
import os
import math

root = os.path.abspath(os.path.join(os.getcwd(), os.pardir))

dice = pd.read_csv(os.path.join(root,"Results","Statistics","dice_stats.csv"))

acc = np.empty((0,len(dice)))
cons = np.empty((0,len(dice)))

for index, row in dice.iterrows():
    filename = str(row["Image"])[:-2]
    if not math.isnan(row["UNet_ROI_FemHead_Pred_Smooth"]):
        acc = np.append(acc,row["UNet_ROI_FemHead_Pred_Smooth"])
#         con_l = rms_contrast(filename+"_l",img_dir="ROI")
#         con_r = rms_contrast(filename+"_r",img_dir="ROI")
#         cons = np.append(cons,(con_l+con_r)/2)
        con = rms_contrast(filename)
        cons = np.append(cons,con)
        
```

```python
from numpy.polynomial.polynomial import polyfit
from scipy.stats import linregress

plt.scatter(cons,acc)
# c, stats = polyfit(cons, acc, 1, full=True)
slope, intercept, r_value, p_value, std_err = linregress(cons, acc)
plt.plot(cons, slope*cons + intercept, color='red', label='Regression Line')
# plt.plot(cons, c[1]*cons + c[0], color='red', label='Regression Line')
plt.legend()
plt.xlabel("Contrast")
plt.ylabel("Dice Score")
plt.ylim([0.4,1.0])

print(r_value**2)
```
